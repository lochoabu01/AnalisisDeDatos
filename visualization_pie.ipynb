```python
# Tema: Comparendos Norte de Santander 2016 - 2020
# Grupo 1: Angie Moreno/202110008171 - Diego Vargas/202110047171 - Luis Ochoa/202519024507
#importamos la librerias para poder trabajar con nuestro DataFrame y poder graficar nuestros graficos de barras
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #Nota: Se importa esta libreria para ajustar propiedades de las graficas. Los graficos en si los estamos haciendo con "seaborn"
from matplotlib.ticker import FuncFormatter # al igual que esta, para poder colocar formato en los ejes
```

```python
# Cargamos el CSV a un DataFrame resultante del proceso "cleaning" (df)
df = pd.read_csv("ComparendosNorteSantanderSource_clean.csv")
```

```python
# Verificamos la cantidad de columnas y filas de nuestro df, para verificar que se halla cargado correctamente
df.shape
```

```python
#Observamos las columnas disponibles en nuestro df
df.info()
```

```python
# Obtenemos la cantidad de registros por CLASE_VEHICULO
df_vehiculos = df['CLASE_VEHICULO'].value_counts().reset_index()
df_vehiculos.columns = ['CLASE_VEHICULO', 'Cantidad']

# Definimos el tamaño de la grafica
plt.figure(figsize=(7, 7))

# Crear grafico de torta
wedges, texts, autotexts = plt.pie(df_vehiculos['Cantidad'], autopct='%1.1f%%', startangle=90) #se le retiro la propiedad de label para usar la leyenda

# Nombrar titulo de la grafica
plt.title('Distribución de infracciones por tipo de vehículo')

# Agregamos leyenda con las categorías, debido a que como etiquetas o label se sobre-montan entre ellas
plt.legend(wedges, df_vehiculos['CLASE_VEHICULO'], title="Tipo de vehiculo", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Mostrar torta
plt.show()

# Donde se evidencia que el 44,4% de los vehiculos infractores son Motocicletas y un 24,5% Automoviles. Frente a los demás vehiculos que representan un menor porcentaje
```

```python
# Extraemos el año de la columna FECHA
df['AÑO'] = pd.to_datetime(df['FECHA']).dt.year

# Obtenemos la cantidad de registros por año
df_anios = df['AÑO'].value_counts().reset_index()
df_anios.columns = ['AÑO', 'Cantidad']

# Definimos el tamaño de la gráfica
plt.figure(figsize=(8, 8))

# Creamos grafico de torta
plt.pie(df_anios['Cantidad'], labels=df_anios['AÑO'], autopct='%1.1f%%', startangle=90)

# Nombramos el titulo de la gráfica
plt.title('Distribución de infracciones por año')

# Mostrar torta
plt.show()

#Donde se puede evidenciar que en el 2016 se cometio el mayor porcentaje de infracciones con un 33,9%; frente a 2019 con el menor porcentaje de infraciones, con un 19,1%
```

```python
# Obtenemos la cantidad de registros por CODIGO_INFRACCION (Top 10)
df_top_infracciones = df['CODIGO_INFRACCION'].value_counts().head(10).reset_index()
df_top_infracciones.columns = ['CODIGO_INFRACCION', 'Cantidad']

# Definimos el tamaño de la gráfica
plt.figure(figsize=(8, 8))

# Creamos grafico de torta 
plt.pie(df_top_infracciones['Cantidad'], autopct='%1.1f%%',startangle=90) #se le retiro la propiedad de label para usar la leyenda

# Agregar leyenda de los codigos de infraccion
plt.legend(df_top_infracciones['CODIGO_INFRACCION'], title="Codigo de infraccion", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Nombramos titulo de la gráfica
plt.title('Top 10 de Infraciones de Transito')

# Mostrar torta
plt.show()

#Donde se puede identificar que dentro del TOP 10 de las infracciones mas cometidas: la infracion D01 con 17,9% y C35 con 17,3% son las infracciones con mayor porcentaje
# de infraccion, las cuales corresponden a: conducir sin licencia de conduccion y no tener vigente la tecnomecanica respectivamente.
```
